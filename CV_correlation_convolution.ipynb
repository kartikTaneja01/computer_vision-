```python
import numpy as np
import matplotlib.pyplot as plt

# Load a sample grayscale image (replace with your own image)
img = plt.imread('d.jpg')


```

```python
# Define a sample 3x3 filter kernel (you can use your own kernel)
kernel = np.array([[-1,-1,-1],
                   [0, 0, 0],
                   [1,1,1]])
```

```python
# Get the dimensions of the image and kernel
M = img.shape[0]
N = img.shape[1]
MN = M*N
MN

```

```python
kernel_height, kernel_width = kernel.shape
```

```python
# Initialize an empty result array for correlation and convolution
correlation_result = np.zeros((M,N))
convolution_result = np.zeros((M,N))
```

```python
# Perform correlation and convolution
for i in range(M - kernel_height + 1):
    for j in range(N - kernel_width + 1):
        image_patch = img[i:i+kernel_height, j:j+kernel_width]
        correlation_result[i, j] = np.sum(image_patch * kernel)
        convolution_result[i, j] = np.sum(image_patch * np.flipud(np.fliplr(kernel)))
```

```python
# Display the original image and results
plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
```

```python
plt.subplot(133)
plt.imshow(correlation_result, cmap='gray')
plt.title('Correlation Result')
```

```python
plt.subplot(132)
plt.imshow(convolution_result, cmap='gray')
plt.title('Convolution Result')

plt.show()
```
