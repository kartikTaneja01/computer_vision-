```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1 =cv2.imread('image5.jpg')
image2 =cv2.imread('image6.png')
```

```python
image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))
```

```python
#BItwise AND operation
from cv2.gapi import bitwise_and
bitwise_and = cv2.bitwise_and(image1, image2)
```

```python
# Bitwise OR operation to combine images
from cv2.gapi import bitwise_or
bitwise_or = cv2.bitwise_or(image1, image2)
```

```python
# Bitwise XOR operation to combine images
from cv2.gapi import bitwise_xor
bitwise_xor = cv2.bitwise_xor(image1, image2)
```

```python
# Bitwise XOR operation to combine images
from cv2.gapi import bitwise_not
bitwise_not = cv2.bitwise_not(image1, image2)
```

```python
from google.colab.patches import cv2_imshow
```

```python
# Display the original and processed images
cv2_imshow(image1)
cv2_imshow(image2)
cv2_imshow(bitwise_and)
cv2_imshow(bitwise_or)
cv2_imshow(bitwise_xor)
cv2_imshow(bitwise_not)
```
