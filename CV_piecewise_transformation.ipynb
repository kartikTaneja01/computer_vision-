```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

# Load the image
image_path = 'dollar.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if image is None:
    print("Error loading image")
    exit()

# Perform bitwise plane slicing
bit_planes = [np.bitwise_and(image, 2 ** i) for i in range(8)]

# Display each bit plane using cv2_imshow
for i, plane in enumerate(bit_planes):
    cv2_imshow(plane * 255)
```

```python
#gray-level slicing

image_path ="dollar.jpg"
image =cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
# Check if the image loaded successfully
if image is None:
    print("Error: Failed to load the image.")

```

```python
# Define lower and upper intensity thresholds
lower_threshold = 100 # Adjust as needed
upper_threshold = 500 # Adjust as needed

# Create a mask to highlight the intensity range between thresholds
mask = cv2.inRange(image, lower_threshold, upper_threshold)

# Apply the mask to the original image to perform gray-level slicing
sliced_image = cv2.bitwise_and(image, image, mask=mask)

# Display the original and sliced images
plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.subplot(132)
plt.imshow(mask, cmap='gray')
plt.title('Mask')
plt.subplot(133)
plt.imshow(sliced_image, cmap='gray')
plt.title('Sliced Image')
plt.show()
```

```python
#thresholding

image_path ="dollar.jpg"
image =cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Set a threshold value (adjust as needed)
threshold_value = 200

# Apply binary thresholding
_, thresholded_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

# Display the original and thresholded images
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.subplot(122)
plt.imshow(thresholded_image, cmap='gray')
plt.title('Thresholded Image')
plt.show()
```
