Average Filtering

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# Read the image
img = cv2.imread('/content/hd.jpg')
```

```python
img
```

```python
m, n, _ = img.shape
```

```python
mask = np.array([[1, 1, 1],
                 [1, 1, 1],
                 [1, 1, 1]])
```

```python
mask = mask/9
```

```python
img_avg = np.zeros_like(img)
```

```python
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = (
            img[i-1, j-1]*mask[0, 0] + img[i-1, j]*mask[0, 1] + img[i-1, j + 1]*mask[0, 2] +
            img[i, j-1]*mask[1, 0] + img[i, j]*mask[1, 1] + img[i, j + 1]*mask[1, 2] +
            img[i + 1, j-1]*mask[2, 0] + img[i + 1, j]*mask[2, 1] + img[i + 1, j + 1]*mask[2, 2]
        )

        img_avg[i, j] = temp
```

```python
img_avg
```

```python
cv2_imshow(img)
cv2_imshow(img_avg)
```

Median

```python
img_med = np.zeros([m, n])
```

```python
mask = np.array([[1, 1, 1],
                 [1, 1, 1],
                 [1, 1, 1]])
```

```python
for i in range(1, m-1):
    for j in range(1, n-1):
      k = np.array([img[i-1, j-1]*mask[0, 0], img[i-1, j]*mask[0, 1], img[i-1, j + 1]*mask[0, 2],
          img[i, j-1]*mask[1, 0], img[i, j]*mask[1, 1], img[i, j + 1]*mask[1, 2],
          img[i + 1, j-1]*mask[2, 0], img[i + 1, j]*mask[2, 1], img[i + 1, j + 1]*mask[2, 2]])

      img_med[i, j] = np.median(k)
```

```python
img_med
```

```python
cv2_imshow(img)
cv2_imshow(img_med)
```

```python
img_min = np.zeros([m, n])
```

```python
for i in range(1, m-1):
    for j in range(1, n-1):
      k = np.array([img[i-1, j-1]*mask[0, 0], img[i-1, j]*mask[0, 1], img[i-1, j + 1]*mask[0, 2],
          img[i, j-1]*mask[1, 0], img[i, j]*mask[1, 1], img[i, j + 1]*mask[1, 2],
          img[i + 1, j-1]*mask[2, 0], img[i + 1, j]*mask[2, 1], img[i + 1, j + 1]*mask[2, 2]])

      img_min[i, j] = np.min(k)
```

```python
img_min
```

```python
cv2_imshow(img)
cv2_imshow(img_min)
```

```python
img_max = np.zeros([m, n])
```

```python
for i in range(1, m-1):
    for j in range(1, n-1):
      k = np.array([img[i-1, j-1]*mask[0, 0], img[i-1, j]*mask[0, 1], img[i-1, j + 1]*mask[0, 2],
          img[i, j-1]*mask[1, 0], img[i, j]*mask[1, 1], img[i, j + 1]*mask[1, 2],
          img[i + 1, j-1]*mask[2, 0], img[i + 1, j]*mask[2, 1], img[i + 1, j + 1]*mask[2, 2]])

      img_max[i, j] = np.max(k)
```

```python
img_max
```

```python
cv2_imshow(img)
cv2_imshow(img_max)
```
